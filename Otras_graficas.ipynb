```python
import matplotlib.pyplot as plt
import math

def f(x):
  return math.sin( 2 * math.pi * x )

n = 10  

X = [ x/(n-1) for x in range(n) ] 
Y = [ f(x) for x in X ]

plt.plot(X, Y)    
plt.show()        
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.arange(0, 2*3.14 , 0.25)
plt.plot(x,np.sin(x),'m*')       
plt.plot(x,np.cos(x),'r--')      
plt.show()
```

```python
x = np.arange(0, 2*3.14, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'b-', label='Seno')   
plt.plot(x, y2, 'r-.', label='Coseno') 
plt.legend() 
plt.show()
```

```python
x = np.arange(0, 2*3.14, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)   
plt.plot(x, y1, 'b-')  

plt.subplot(2, 1, 2)   
plt.plot(x, y2, 'r-.') 

plt.tight_layout()     

print('Ejemplo 1:\n')
plt.show()

plt.subplot(1, 2, 1)   
plt.plot(x, y1, 'b-')  

plt.subplot(1, 2, 2)   
plt.plot(x, y2, 'r-.') 

plt.tight_layout()     

print('\nEjemplo 2:\n')
plt.show()
```
